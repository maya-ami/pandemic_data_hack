```python
import pandas as pd
import numpy as np
import requests
import os
import sys
# from selenium import webdriver
# from selenium.webdriver.support.ui import WebDriverWait
# from selenium.webdriver.common.by import By
# from selenium.webdriver.support import expected_conditions as EC
# from selenium.common.exceptions import TimeoutException
import time
from bs4 import BeautifulSoup
# from selenium.webdriver.common.action_chains import ActionChains
# from selenium.webdriver.common.keys import Keys
```

```python
page = requests.get('https://russian-trade.com/coronavirus/russia/')
```

```python
soup = BeautifulSoup(page.text, 'html.parser')
```

```python
table = soup.find('span', attrs={'class':'sal-agg-nonbase__average-salary-value'})
```

```python
trs = soup.find_all('curs_special')
table = soup.find('table', attrs={'id':'curs_special'})
```

```python
rows = table.find_all('tr')
```

```python
rows
```

```python
data = []
for row in rows:
    cols = row.find_all('td')

#     if len(cols)< 5:
#         print(cols)
    cols = [ele.text.strip() if len(ele) else '' for ele in cols ]
    data.append(cols)
```

```python
data = pd.DataFrame(data[1:],columns = ['Дата','Заражений','Смертей','Выздоровлений','Заражено на дату'])
```

```python
data.to_csv('coronavir_stat.csv',index=False)
```

```python
data.university_status = data.university_status.fillna('отсутствует')
```

```python
data.drop('university_id',axis=1,inplace=True)
```

```python
data.to_csv('unversity_balls.csv')
```
