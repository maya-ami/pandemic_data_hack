```python
train['position_lower'] = train['position'].apply(lambda x: x.lower())
```

```python
test['position_lower'] = test['position'].apply(lambda x: x.lower())
```

```python
def not_one_position(line):
  if ',' in line:
    return line.split(',')[0]
  else:
    return line
```

```python
train['one_position'] = train['position_lower'].apply(not_one_position)
```

```python
test['one_position'] = test['position_lower'].apply(not_one_position)
```

```python
def larger_position(line):
  if line !='':
    if 'начальник' in line or 'директор' in line or 'руководитель' in line:
      return 'начальник'

    elif 'охран' in line or 'сторож' in line:
      return 'охранник'

    elif 'учитель' in line or 'педагог' in line or 'преподаватель' in line:
      return 'продавец'

    elif 'продавец' in line or 'продаж' in line:
      return 'продавец'

    elif 'юрис' in line:
      return 'юрист'

    elif 'водитель' in line:
      return 'водитель'

    elif 'рабочий' in line or 'разнорабочий' in line or 'подсобный' in line:
      return 'рабочий'

    elif 'туризм' in line or 'турист' in line:
      return 'турагент'

    elif 'экономи' in line:
      return 'экономист'

    elif 'маркет' in line:
      return 'маркетолог'

    elif 'машинист' in line or 'машиннист' in line:
      return 'машинист'

    elif 'программист' in line or 'програмист' in line or 'сисадмин' in line or 'сис.админ' in line or 'системный администратор' in line or '1с' in line or '1c' in line:
      return 'программист'

    elif 'hr' in line or 'кадр' in line or 'персоналу' in line:
      return 'кадровик'

    elif 'аналитик'in line or 'анализ' in line or 'data' in line:
      return 'аналитик'

    elif 'мастер' in line:
      return 'мастер'

    elif 'дизайн' in line:
      return 'дизайнер'

    elif 'инженер' in line:
      return 'инженер'
    
    elif 'касс' in line:
      return 'кассир'
      
    elif 'воспитатель' in line:
      return 'воспитатель'
    
    elif 'нефт' in line:
      return 'нефтяник'

    elif 'геолог' in line:
      return 'геолог'
      
    elif 'хими' in line:
      return 'химик'

    elif 'переводчик' in line:
      return 'переводчик'

    elif 'ветеринар' in line or 'зоо' in line:
      return 'ветеринар'

    elif 'сестра' in line:
      return 'медсестра'

    elif 'фельдшер' in line:
      return 'фельдшер'

    elif 'врач' in line:
      return 'врач'


    elif 'механик' in line:
      return 'механик'

    elif 'банк' in line or 'кредит' in line or 'ипоте' in line:
      return 'врач'

    elif 'психолог' in line:
      return 'психолог'

    elif 'повар' in line or 'пекарь' in line or 'кондитер' in line:
      return 'повар'

    elif 'оператор пк' in line or 'оператор эвм' in line:
      return 'оператор пк'

    elif 'сантехник' in line:
      return 'сантехник'

    elif 'слесарь' in line:
      return 'слесарь'

    elif 'тракторист' in line:
      return 'тракторист'

    elif 'бухгалтер' in line:
      return 'бухгалтер'

    elif 'контролер' in line or 'контролёр' in line:
      return 'слесарь'

    elif 'монтаж' in line:
      return 'монтажник'

    elif 'электромонтер' in line:
      return 'электромонтер'

    elif 'бригадир' in line or 'прораб' in line:
      return 'бригадир'

    elif 'строит' in line or 'стройк' in line:
      return 'строитель'
    
    elif 'секретарь' in line:
      return 'секретарь'

    elif 'тренер' in line:
      return 'тренер'

    elif 'социальн' in line:
      return 'соцработник'

    elif 'чиновн' in line or 'госсуж' in line or 'государств' in line or 'гос'in line:
      return 'госслужащий'

    elif 'уборщ' in line or 'уборк' in line:
      return 'уборщица'
  

    else:
      return line

  else:
    return line
```
